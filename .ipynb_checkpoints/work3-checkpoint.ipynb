```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
img = cv2.imread('data/paimeng.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

height = img.shape[0]
width = img.shape[1]

plt.figure()
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 傅里叶变换
参考：https://blog.csdn.net/qq_36554582/article/details/88701865 &nbsp;
https://stackoverflow.com/questions/60073076/how-to-find-phase-spectrum-of-an-image-in-opencv-python

```python
fft = np.fft.fft2(gray)
fftshift = np.fft.fftshift(fft)
```

### 幅度谱，相位谱

```python
f = np.log(np.abs(fftshift))
p = np.angle(fftshift)
plt.figure(figsize=(18,8))
plt.subplot(131)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(132)
plt.imshow(f, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(133)
plt.imshow(p, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 幅度谱，相位谱恢复

```python
fftifft = np.real(np.fft.ifft2(np.abs(fft)*np.exp(1j*(np.angle(fft)))))

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(fftifft, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 幅度谱，相位谱=0

```python
fftifft = np.real(np.fft.ifft2(np.abs(fft)))
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(fftifft, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 幅度谱=1，相位谱

```python
fftifft = np.real(np.fft.ifft2(np.exp(1j*np.angle(fft))))
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(fftifft, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```
