```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('data/paimeng.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

height = img.shape[0]
width = img.shape[1]

plt.figure()
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 线性点运算
s = ar + b
### 灰度上移（b=50, a=1)

```python
result = np.zeros((height, width), np.uint8)

# 图像灰度反转
for i in range(height):
    for j in range(width):
        if (gray[i, j])+50>255:
            v = 255
        else:
            v = (gray[i, j])+50
        result[i, j] = np.uint8(v)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(result, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

### 灰度扩展（a=1.5,b=0）

```python
result = np.zeros((height, width), np.uint8)

# 图像灰度反转
for i in range(height):
    for j in range(width):
        if (gray[i, j]*1.5)>255:
            v = 255
        else:
            v = (gray[i, j])*1.5
        result[i, j] = np.uint8(v)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(result, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

### 灰度压缩（a=0.1，b=0）

```python
result = np.zeros((height, width), np.uint8)

# 图像灰度反转
for i in range(height):
    for j in range(width):
        v = (gray[i, j])*0.1
        result[i, j] = np.uint8(v)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(result, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

### 灰度反转 （a=-1, b=0）

```python
result = np.zeros((height, width), np.uint8)

for i in range(height):
    for j in range(width):
        v = -(gray[i, j])+255
        result[i, j] = np.uint8(v)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(result, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 分段线性变化

### (a = 100, b = 200, c = 50, d = 250)

```python
result = np.zeros((height, width), np.uint8)
for x in range(1,height):
    for y in range(1,width):
        if(gray[x,y] <= 100):  
            result[x,y] = int(gray[x,y] *50.0 / 100)
        else:
            if(gray[x,y] > 100 and gray[x,y] <= 200):
                result[x,y] = int((200 / 100.0) * (gray[x,y] - 100) + 50)  
            else:
                result[x,y] = int((5 / 50.0) * (gray[x,y] - 200) + 250)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(result, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 非线性变换
### 幂次变换
### $s = rc^y$ 
### ($y=0.5, c=1$)

```python
result = np.zeros((height, width), np.uint8)
for x in range(1,height):
    for y in range(1,width):
        result[x,y] = int(gray[x,y]**0.5)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(result, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

### ($y=1.1, c=1$)

```python
result = np.zeros((height, width), np.uint8)
for x in range(1,height):
    for y in range(1,width):
        result[x,y] = int(gray[x,y]**1.1)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(result, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

### log变换
### $ s = c log(1 + r)$
### $c = 1$

```python
result = np.zeros((height, width), np.uint8)
for x in range(1,height):
    for y in range(1,width):
        result[x,y] = int(np.log(gray[x,y]))

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(result, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```
