```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
img = cv2.imread('data/paimeng.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

height = img.shape[0]
width = img.shape[1]

plt.figure()
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 傅里叶变换
参考：https://blog.csdn.net/qq_36554582/article/details/88701865

```python
fft = np.fft.fft2(gray)
fftshift = np.fft.fftshift(fft)
```

### 幅度谱，相位谱

```python
# Picture_AM_Spectrum = log(abs(Picture_FFT_Shift));%获得傅里叶变换的幅度谱
# Picture_Phase_Specture = log(angle(Picture_FFT_Shift)*180/pi);%获得傅里叶变换的相位谱
f = np.log(np.abs(fftshift))
p = np.log(np.angle(fftshift)*180/math.pi)
plt.figure(figsize=(12,6))
plt.subplot(131)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(132)
plt.imshow(f, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(133)
plt.imshow(p, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 幅度谱，相位谱恢复

```python
fftifft = np.real(np.fft.ifft2(np.abs(fft)*np.exp(1j*(np.angle(fft)))))

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(fftifft, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 幅度谱恢复

```python
fftifft = np.real(np.fft.ifft2(np.abs(fft)))
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(fftifft, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

## 相位谱恢复

```python
fftifft = np.real(np.fft.ifft2(np.exp(1j*np.angle(fft))))
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray, "gray")
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(fftifft, "gray")
plt.xticks([]), plt.yticks([])
plt.show()
```
